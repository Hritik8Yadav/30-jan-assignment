```python
# Ans1: 
percent=int(input("Enter the percentage: "))
if(percent>90):
    print("A")
elif(percent<=90 and percent>80):
    print("B")
elif(percent<=80 and percent>=60):
    print("C")
elif(percent<60):
    print("D")
else:
    print("wrong input")
```

```python
#Ans 5: while loop is used to do given number of steps repetatively until certain condition is true. 
# unlike in for loop the while loop doesnot require the number of iteration but it execute untill we get the condition false.
# while loop require the termination condition for exit for example: 
a=0
while a<5:
    print(a)
    a+=1
```

```python
a=0
b=0
while(a<5):
    while(b<5):
        print("*"*b)
        b+=1
    a+=1
```

```python
a=5
while(a>0):
    b=a
    while(b>0):
        print(a,end="")
        b-=1
    print()
    a=a-1
```

```python
a=1
c=0
while(c<5):
    while(a<=5):
        print(" "*(5-a),end="")
        print("*"*a,end="")
        print()
        a+=1
    c+=1
```
